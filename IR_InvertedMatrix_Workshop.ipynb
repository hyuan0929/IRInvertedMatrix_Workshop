```python
import os
import requests
import feedparser

# Example: Download blog posts from a public RSS feed and save as .txt files
def fetch_and_save_blog_posts(rss_url, save_folder, max_posts=20):
    feed = feedparser.parse(rss_url)
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)
    count = 0
    for entry in feed.entries:
        if count >= max_posts:
            break
        title = entry.get('title', 'untitled')
        content = entry.get('summary', '')
        # Clean filename
        filename = f"blog_{count+1}.txt"
        filepath = os.path.join(save_folder, filename)
        with open(filepath, 'w', encoding='utf-8') as f:
            f.write(f"{title}\n\n{content}")
        count += 1
    print(f"Saved {count} blog posts to {save_folder}")

# Example RSS feed: Python Software Foundation blog
rss_url = 'https://pyfound.blogspot.com/feeds/posts/default?alt=rss'
fetch_and_save_blog_posts(rss_url, 'sample_docs', max_posts=20)
```

```python
# Example: Load text files from a folder
import os

def load_documents(folder_path):
    documents = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
                documents.append(file.read())
    return documents

# Replace 'sample_docs/' with your actual folder
documents = load_documents('sample_docs/')
print(f"Loaded {len(documents)} documents.")
```

## ✂️ Step 2: Tokenizer


### 🗣 Instructor Talking Point:
> The tokenizer breaks raw text into a stream of words (tokens). This is the foundation for every later step in IR and NLP.

### 🔧 Your Task:

```python
import re

def tokenize(text):
    tokens = re.findall(r'\b\w+\b', text.lower())
    return tokens

# Test on one document
tokens = tokenize(documents[0])
print(tokens[:20])  # Preview first 20 tokens
```

## 📝 Team Talking Point – Tokenizer

In this step, we used a basic tokenizer to split raw text into lowercase word tokens. The tokenizer applies a regular expression to extract word-like patterns and removes most punctuation automatically.

This step is important because tokenization converts unstructured text into a format that can be processed in later stages such as normalization and inverted indexing.

A practical issue we observed is that simple tokenization may still keep noisy terms from URLs or HTML-like fragments. However, this lightweight approach is still effective as a first step in the NLP pipeline.

## 🔁 Step 3: Normalization Pipeline (Stemming, Stop Word Removal, etc.)


### 🗣 Instructor Talking Point:
> Now we normalize tokens: convert to lowercase, remove stop words, apply stemming or affix stripping. This reduces redundancy and enhances search accuracy.

### 🔧 Your Task:
- Use `nltk` to remove stopwords and apply stemming.


```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def normalize_tokens(tokens):
    return [stemmer.stem(t) for t in tokens if t not in stop_words]

# Example: normalize one document
norm_tokens = normalize_tokens(tokens)
print(norm_tokens[:20])
```

## 📝 Team Talking Point – Normalization Pipeline

In this step, we applied normalization techniques including stop word removal and stemming to clean and standardize the tokens.

Stop words such as "the", "is", and "and" were removed because they do not carry significant meaning for search or indexing. This helps reduce noise in the dataset.

We also applied stemming using the PorterStemmer to reduce words to their root form. For example, words like "running", "runs", and "ran" can be reduced to a common base form. This improves matching during search queries.

One challenge we observed is that stemming can sometimes produce non-intuitive words (e.g., "applications" becomes "applic"), but it is still useful for improving recall in search systems.

## 🔍 Step 4: Inverted Index


### 🗣 Instructor Talking Point:
> We now map each normalized token to the list of document IDs in which it appears. This is the core structure that allows fast Boolean and phrase queries.

### 🔧 Your Task:
- Build the inverted index using a dictionary.
- Add code to support phrase queries using positional indexing.


```python
from collections import defaultdict

def build_inverted_index(documents):
    index = defaultdict(list)
    for doc_id, text in enumerate(documents):
        tokens = normalize_tokens(tokenize(text))
        seen = set()
        for token in tokens:
            if token not in seen:
                index[token].append(doc_id)
                seen.add(token)
    return dict(index)

inverted_index = build_inverted_index(documents)
print(dict(list(inverted_index.items())[:10]))  # Preview first 10 terms
```

## 📝 Team Talking Point – Inverted Index

In this step, we built an inverted index that maps each token to a list of document IDs where the token appears. This structure is essential for efficient information retrieval.

By using a dictionary, we can quickly look up which documents contain a given word. This significantly improves search performance compared to scanning all documents.

We also ensured that each document ID appears only once per token by using a set to track seen tokens. This avoids redundancy in the index.

One limitation of this basic inverted index is that it does not store word positions, which makes phrase queries difficult. This motivates the use of positional indexing for more advanced search capabilities.

```python
# Placeholder for phrase query implementation
# You may build a position-aware index or use string search within docs after normalization

# Example:
query1 = "Python software"
query2 = "bylaws"

def phrase_query(query, documents):
    results = []
    query_tokens = normalize_tokens(tokenize(query))
    for doc_id, doc in enumerate(documents):
        doc_tokens = normalize_tokens(tokenize(doc))
        # Search for the sequence in the token list
        for i in range(len(doc_tokens) - len(query_tokens) + 1):
            if doc_tokens[i:i+len(query_tokens)] == query_tokens:
                results.append(doc_id)
                break
    return results

# Run phrase queries on the loaded documents
results1 = phrase_query(query1, documents)
results2 = phrase_query(query2, documents)

print(f"Documents containing the phrase '{query1}': {results1}")
print(f"Documents containing the phrase '{query2}': {results2}")
```

## 📝 Team Talking Point – Phrase Queries

In this step, we implemented phrase queries to search for exact sequences of words within documents. Unlike single-term queries, phrase queries require that words appear in the correct order and consecutively.

We used a sliding window approach to compare sequences of tokens in each document with the query tokens. This allows us to identify documents that contain the exact phrase.

One limitation of this approach is that it scans each document sequentially, which can be inefficient for large datasets. A more advanced solution would involve positional indexing to improve performance.

```python
# Implement Positional Indexes: Store term positions for each document
from collections import defaultdict

def build_positional_index(documents):
    positional_index = defaultdict(lambda: defaultdict(list))
    for doc_id, text in enumerate(documents):
        tokens = normalize_tokens(tokenize(text))
        for pos, token in enumerate(tokens):
            positional_index[token][doc_id].append(pos)
    return positional_index

# Build positional index for the loaded documents
positional_index = build_positional_index(documents)

# Example: Show positions for a sample term
sample_term = 'python'  # Change to any term you want to inspect
print(f"Positions for term '{sample_term}':")
for doc_id, positions in positional_index[sample_term].items():
    print(f"  Document {doc_id}: {positions}")
```

## 📝 Team Talking Point – Bigram Search

For searching bigrams (pairs of consecutive words), we prefer using a positional index rather than a simple inverted index or term-document count matrix.

A positional index stores the exact positions of terms in each document, which allows us to verify whether two words appear next to each other in the correct order. This makes phrase and bigram queries more accurate.

Although positional indexes require more storage and processing time, they provide significantly better precision for phrase-based search tasks compared to basic indexing methods.

### 📊 Comparison: Positional Index vs. Term Document Count Matrix

A **Positional Index** stores not only which documents a term appears in, but also the exact positions of each term within those documents. In contrast, a **Term Document Count Matrix** (TDCM) simply records the frequency of each term in each document, without any positional information.

| Feature                      | Positional Index                | Term Document Count Matrix (TDCM) |
|------------------------------|---------------------------------|-----------------------------------|
| Stores term positions        | Yes                             | No                                |
| Supports phrase/bigram search| Yes                             | No                                |
| Memory usage                 | Higher                          | Lower                             |
| Query speed for phrases      | Fast                            | Slow (requires scanning)          |
| Useful for                   | Phrase queries, proximity search | Keyword frequency analysis         |
| Implementation complexity    | Higher                          | Lower                             |

**Summary:**
- Use a positional index for advanced search features like phrase and proximity queries.
- Use a TDCM for simple keyword frequency analysis and ranking.

## 📝 Team Talking Point – IDF (Inverse Document Frequency)

In this step, we implemented the IDF calculation to measure how rare a term is across the document collection.

IDF assigns higher values to terms that appear in fewer documents, making them more important for distinguishing relevant content. Common words that appear in many documents receive lower IDF scores.

We used the formula:

IDF = log(total_documents / document_frequency)

This improves search relevance when combined with TF in TF-IDF. However, IDF may overemphasize rare but unimportant terms and requires knowledge of the entire dataset.

### 📐 Mathematical Explanation: Inverse Document Frequency (IDF)

The **Inverse Document Frequency (IDF)** is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. The intuition is that terms that appear in many documents are less informative than those that appear in few.

The IDF for a term $t$ is defined as:

$$
\text{IDF}(t) = \log\left(\frac{N}{n_t}\right)
$$

Where:
- $N$ is the total number of documents in the corpus.
- $n_t$ is the number of documents containing the term $t$.

---

#### Example Calculation

Suppose:
- $N = 20$ (total documents)
- $n_{\text{softwar}} = 15$ (documents containing the term 'softwar')

Then:

$$
\text{IDF}(\text{softwar}) = \log\left(\frac{20}{15}\right) \approx 0.2877
$$

So, the IDF for term 'softwar' is $0.2877$.

A higher IDF value means the term is rare across the corpus, while a lower value means the term is common. IDF is often used in combination with Term Frequency (TF) to compute the TF-IDF score:

$$
\text{TF-IDF}(t, d) = \text{TF}(t, d) \times \text{IDF}(t)
$$

Where $\text{TF}(t, d)$ is the frequency of term $t$ in document $d$.

This weighting helps highlight terms that are both frequent in a document and rare across the corpus, improving search relevance and document ranking.

```python
# Calculate Term Frequency (TF) for the term 'softwar' in each document
# TF is simply the number of times the term appears in a document
import pandas as pd

term = 'softwar'  # Use the stemmed term as before

tf_values = []
for doc_id, doc in enumerate(documents):
    tokens = normalize_tokens(tokenize(doc))
    tf = tokens.count(term)
    tf_values.append(tf)
    print(f"TF for term '{term}' in Document {doc_id}: {tf}")

# Display as a DataFrame for clarity
TF_df = pd.DataFrame({'Document': list(range(len(documents))), 'TF': tf_values})
print(TF_df)
```

## 📝 Team Talking Point – TF-IDF Example

From the output above, we can observe that the term "softwar" appears **6 times** in Document 0.

Previously, we computed the IDF value for "softwar" as approximately **0.2231**.

Using the TF-IDF formula:

TF-IDF = TF × IDF

For Document 0:

TF-IDF = 6 × 0.2231 ≈ 1.3386

This shows that although "softwar" appears frequently in Document 0, its overall importance is moderated by its presence across multiple documents.

TF-IDF helps balance term frequency with term rarity, making it useful for identifying important terms for search and ranking tasks.

```python
def query_processor_tfidf(query, documents, idf_scores):
    results = []

    for doc_id, doc in enumerate(documents):
        tokens = normalize_tokens(tokenize(doc))
        score = 0

        for term in query.split():
            tf = tokens.count(term)
            idf = idf_scores.get(term, 0)
            score += tf * idf

        if score > 0:
            results.append((doc_id, score))

    # sort by score (descending)
    results.sort(key=lambda x: x[1], reverse=True)

    return results
```

## 📝 Reflection

In this section, we extended a basic query processor to incorporate TF-IDF scoring for ranking documents.

The original approach only returned documents containing the query terms, without considering their importance. By using TF-IDF, we can rank documents based on both term frequency and term rarity.

This improves search relevance by prioritizing documents where the query terms are both frequent and distinctive.

One limitation of this approach is that it still treats terms independently and does not consider word order or phrase structure, which can be addressed using positional indexes.
